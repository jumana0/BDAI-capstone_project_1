## Content Based

```python
linear_kernel_Dec_matrix = linear_kernel(Dec_matrix, Dec_matrix) #Compute the similarity for the description matrix
linear_kernel_porgram_dec_matrix = linear_kernel(porgram_dec_matrix, porgram_dec_matrix) #Compute the similarity for the program description matrix
linear_kernel_full_Dec_matrix = linear_kernel(full_Dec_matrix, full_Dec_matrix) #Compute the similarity for thr full description matrix
```

```python
name = Dec_df['name'] # unique movies name
indices = pd.Series(Dec_df.index, index=name) #series of indexes

def recommendations_porgram_dec(movie):
    idx = indices[movie] #get the name index
    sim_scores = list(enumerate(linear_kernel_porgram_dec_matrix[idx])) #list of similarity
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)  #sort the list from the higher probability
    sim_scores = sim_scores[1:4] #choose the values from 1 to 4
    movie_indices = [i[0] for i in sim_scores] #get the movies names and store it.
    return name.iloc[movie_indices].tolist() 
```

```python
def recommendations_Dec(movie):
    idx = indices[movie] #get the name index
    sim_scores = list(enumerate(linear_kernel_Dec_matrix[idx]))  #list of similarity
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True) #sort the list from the higher probability
    sim_scores = sim_scores[1:4]  #choose the values from 1 to 4
    movie_indices = [i[0] for i in sim_scores]  #get the movies names and store it.
    return name.iloc[movie_indices].tolist() 
```

```python
def recommendations_full_Dec(movie):
    idx = indices[movie] #get the name index
    sim_scores = list(enumerate(linear_kernel_full_Dec_matrix[idx]))  #list of similarity
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True) #sort the list from the higher probability
    sim_scores = sim_scores[1:4] #choose the values from 1 to 4
    movie_indices = [i[0] for i in sim_scores] #get the movies names and store it.
    return name.iloc[movie_indices].tolist() 
```

## Collaborative Filtering

```python
collf = pd.read_csv('D:\Bootcamp\Week8\MLCapstone Project (2)\MLCapstone Project\collf.csv') #read collf dataset
collf.drop('Unnamed: 0',axis=1,inplace=True) #drop Unnamed: 0 column
collf.head(5)
```

```python
#create pivot_table, in the rows indexes there is user_id_maped and in the column the movies names filled by rating
rating = collf.pivot_table(index=['user_id_maped'],columns=['name'],values='rating') 
rating = rating.fillna(0,axis=1) #fill nan values by 0
```

```python
matrix = rating.corr(method='pearson') #compute pairwise correlation of columns,by using standard correlation coefficient
matrix.head(5)
```

```python
def recommended(name):
    sim_scores = matrix[name] #get the name with probability
    sim_scores = sim_scores.sort_values(ascending=False) #sort the list from the higher probability
    return sim_scores[1:4].index.tolist() #choose the values from 1 to 4
```
